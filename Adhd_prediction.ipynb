```python
from google .colab import files
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
from sklearn.model_selection import GridSearchCV
```

```python
df = pd.read_csv("adhd_data_new.csv")
df.head()
```

```python
df.info()
```

```python
print("Number of missing value of each column ")
print(df.isnull().sum())
```

```python
print("\n",df["Family_History"].value_counts())
print("\n",df["Educational_Level"].value_counts())
```

```python
df["Diagnosis_Class"].value_counts()
```

```python
conditions = (
    ((df["Focus_Score_Video"] < 3) |
     (df["Difficulty_Organizing_Tasks"] == 1) &
     (df["Daily_Activity_Hours"] > 5)) |
     (df["Learning_Difficulties"] == 1) &
    (df["Diagnosis_Class"] > 1)
)
df["Adhd"] = conditions.astype(int)
df.head()
```

```python
print(df["Adhd"].value_counts())
```

```python
le = LabelEncoder()
df['Family_History'] = le.fit_transform(df["Family_History"])
df.head()
```

```python
encoder = OneHotEncoder(sparse_output=False, drop='first')
encoder_array = encoder.fit_transform(df[['Educational_Level']])
feature_names = encoder.get_feature_names_out(['Educational_Level'])
encoded_df = pd.DataFrame(encoder_array, columns=feature_names)
df = pd.concat([df,encoded_df], axis=1)
```

```python
df=df.drop('Educational_Level',axis=1)
print(df.isnull().sum())
df.head()
```

```python
print(df.corr()['Adhd'].sort_values(ascending=False))
```

```python
Features =df.drop("Adhd", axis=1)
y=df["Adhd"]
```

```python
scaler = StandardScaler()
scaled_df = scaler.fit_transform(Features)
x= pd.DataFrame(scaled_df, columns=Features.columns)
x.head()
```

```python
xtrain,xtest, ytrain,ytest = train_test_split(x,y , test_size = 0.2, random_state=7, stratify = y)
print("Training Feature data size : ",xtrain.shape)
print("Training Target data size : ",ytrain.shape)
print("Testing Feature data size : ",xtest.shape)
print("Testing Target data size : ",ytest.shape)
```

```python
lr=LogisticRegression()
lr.fit(xtrain,ytrain)
ypred = lr.predict(xtest)
accuracy=accuracy_score(ytest,ypred)
print("\n\t\t\t Logistic Regression \n")
print("Accuracy : ",round(accuracy,2))
print("Classification Report : \n", classification_report(ytest,ypred))
```

```python
rf=RandomForestClassifier()
rf.fit(xtrain,ytrain)
ypred = rf.predict(xtest)
accuracy=accuracy_score(ytest,ypred)
print("\n\t\t\t Random Forest Classifier \n")
print("Accuracy : ",round(accuracy,2))
print("Classification Report : \n", classification_report(ytest,ypred))
```

```python
svc=SVC()
svc.fit(xtrain,ytrain)
ypred = svc.predict(xtest)
accuracy=accuracy_score(ytest,ypred)
print("\n\t\t\t Support Vector Classifier \n")
print("Accuracy : ",round(accuracy,2))
print("Classification Report : \n", classification_report(ytest,ypred))
```

```python
kn = KNeighborsClassifier()
kn.fit(xtrain,ytrain)
ypred=kn.predict(xtest)
accuracy = accuracy_score(ytest,ypred)
print("\n\t\t\t KNeighbors Classifier \n")
print("Accuracy : ",round(accuracy,2))
print("Classification Report : \n", classification_report(ytest,ypred))
```

```python
dt = DecisionTreeClassifier()
dt.fit(xtrain,ytrain)
ypred=dt.predict(xtest)
accuracy = accuracy_score(ytest,ypred)
print("\n\t\t\t Decision Tree Classifier \n")
print("Accuracy : ",round(accuracy,2))
print("Classification Report : \n", classification_report(ytest,ypred))
```
