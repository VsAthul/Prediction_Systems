```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder, OneHotEncoder
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
data["Anxiety"] = data.apply(lambda x: 1 if x['Anxiety Level (1-10)'] >=5 else 0, axis=1)
data.head(5)
```

```python
data["Anxiety"].value_counts()
```

```python
data.head()
```

```python
binary_columns = ['Gender','Smoking', 'Family History of Anxiety', 'Dizziness',
                 'Medication', 'Recent Major Life Event']
le = LabelEncoder()
for col in binary_columns:
  data[col]=le.fit_transform(data[col])
occupation_dummies = pd.get_dummies(data['Occupation'], prefix="Occupation")
data=pd.concat([data, occupation_dummies], axis=1)
data=data.drop("Occupation", axis=1)
print(data.head())
print(data.columns.tolist())
```

```python
data.info()
```

```python
print(data.corr()['Anxiety'].sort_values(ascending = False))
```

```python
y=data["Anxiety"]
features=data.drop(['Anxiety','Anxiety Level (1-10)'], axis=1)
print("Target shape : ",y.shape)
print("Feature shape : ",features.shape)
```

```python
sc = StandardScaler()
x=sc.fit_transform(features)
```

```python
x=pd.DataFrame(x, columns=features.columns)
print(x.columns.tolist())
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size= 0.2, stratify=y)
print("training data size : ",xtrain.shape)
print("test data size : ",xtest.shape)
```

```python
lr = LogisticRegression()
lr.fit(xtrain,ytrain)
ypred=lr.predict(xtest)
accuracy=accuracy_score(ytest,ypred)
print("Accuracy score for Logistic Regression : ",round(accuracy,2))
print("Confusion matrix : \n", confusion_matrix(ytest,ypred))
print("Classification Report : \n", classification_report(ytest,ypred))
```

```python
svc = SVC(kernel='linear')
svc.fit(xtrain,ytrain)
ypred=svc.predict(xtest)
accuracy=accuracy_score(ytest,ypred)
print("Accuracy score for Support Vector Classifier : ",round(accuracy, 2))
print("Confusion matrix : \n", confusion_matrix(ytest,ypred))
print("Classification Report : \n", classification_report(ytest,ypred))
```

```python
nb = GaussianNB()
nb.fit(xtrain,ytrain)
ypred=nb.predict(xtest)
accuracy=accuracy_score(ytest,ypred)
print("Accuracy score for Naive byes Classifier : ",round(accuracy, 2))
print("Confusion matrix : \n", confusion_matrix(ytest,ypred))
print("Classification Report : \n", classification_report(ytest,ypred))
```

```python
rf = RandomForestClassifier(n_estimators=100,
    max_depth=10,
    random_state=42)
rf.fit(xtrain,ytrain)
ypred=rf.predict(xtest)
accuracy=accuracy_score(ytest,ypred)
print("Accuracy score for Random Forest Classifier : ",round(accuracy, 2))
print("Confusion matrix : \n", confusion_matrix(ytest,ypred))
print("Classification Report : \n", classification_report(ytest,ypred))
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(xtrain,ytrain)
ypred=knn.predict(xtest)
accuracy=accuracy_score(ytest,ypred)
print("Accuracy score for KNeighbors Classifier : ",round(accuracy, 2))
print("Confusion matrix : \n", confusion_matrix(ytest,ypred))
print("Classification Report : \n", classification_report(ytest,ypred))
```

To improve the accuracy and model performance of KNeighbors classifier model

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={
    'n_neighbors': range(3,20),
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan', 'minkowski']
}
knngrid = GridSearchCV(estimator=knn, param_grid = param_grid, cv =5, scoring='accuracy', verbose=1)
knngrid.fit(xtrain,ytrain)
```

```python
ypred = knngrid.predict(xtest)
accuracy=accuracy_score(ytest,ypred)
print("Accuracy score for KNeighbors Classifier : ",round(accuracy, 2))
print("Confusion matrix : \n", confusion_matrix(ytest,ypred))
print("Classification Report : \n", classification_report(ytest,ypred))
```
